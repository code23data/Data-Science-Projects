## Removing the "education" Column

Since the "education" column is not a relevant feature for the model that we want to build, we are removing that column from the entire dataset.

# Data Preprocessing

## Checking for Null Values

```python
new_df.isnull().sum()
```

## Removing Rows with `NaN` (null) Values

```python
new_df = new_df.dropna(how='any', axis=0)
```

```python
# Validating if there are still any remaining NaN values in the dataset
new_df.isnull().sum()
```

## Separating the Features and Target

Now, we shall separate the features from the target column. The target column being `TenYearCHD`, which is what we want out model to predict on test data.

```python
# Separating the features from the target
X, y = new_df.drop('TenYearCHD', axis=1), new_df['TenYearCHD']

# Checking if both the features and target have the same number of rows.
X.shape, y.shape
```

## Splitting the Dataset 

Now, to do further prerpocessing and model training, we need to split our dataset into training and testing datasets respectively. For that, we will use the `train_test_split` function of the Scikit-Learn library. Splitting the dataset is important to make sure that the predictions made by the final model on the test data is actually learnt and not "memorized" by the dataset provided.

We will split the dataset in such a manner that 30% of the dataset will belong to the test dataset, and the remaining 70% will belong to the training dataset.

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into train(70%) and test(30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Viewing the first 5 rows of training dataset
X_train.head()
```

## Normalizing Numerical Features

Now we shall normalize the numerical features using `StandardScaler` from Sci-kit Learn library, for uniform scaling. However, we need to first separate out the numerical features from the categorical features. The categorical featuers are: `male`, `currentSmoker`, `prevalentStroke`, `prevalentHyp`, `diabetes`. The remaining are numerical features that need to be scaled or normalized.

```python
# Importing the StandardScaler from sklearn library.
from sklearn.preprocessing import StandardScaler

# Creating an instance of StandardScaler
scaler = StandardScaler()

# Fitting and transforming the scaler on the training data
X_train[['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']] = scaler.fit_transform(X_train[['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']])

# Normalizing the test dataset
X_test[['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']] = scaler.transform(X_test[['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']])

X_train.head()
```

## Accuracy Score
For binary classification models, if $\hat{y_i}$ is the predicted value of the $i$-th sample and $y_i$ is the corresponding true value, then the fraction of correct predictions over $n_{samples}$ is defined as:
$$
\text{accuracy}(y, \hat{y})=\frac{1}{n_{samples}}\sum_{i=0}^{n_{samples}-1}\mathbb{1}(\hat{y_i}=y_i)
$$
Here, $\mathbb{1}(x)$ is the indicator function, which can be defined as:
$$
\mathbb{1}(x) = \begin{cases}
                1 & x \text{ is True}\\
                0 & \text{otherwise}
                \end{cases}
$$

```python
# Importing the accuracy_score function
from sklearn.metrics import accuracy_score

# Evaluating accuracy score for logistic regression model
accuracy_logreg = accuracy_score(y_test, y_pred1)

# Evaluating accuracy score for random forest classifier model
accuracy_random_forest = accuracy_score(y_test, y_pred2)

# Evaluating accuracy score for grid search random forest classifier model
accuracy_random_forest_cv = accuracy_score(y_test, y_pred3)

# Evaluating accuracy score for xgboost model
accuracy_xgboost = accuracy_score(y_test, y_pred4)

print(f"The accuracy score for logistic regression model is {accuracy_logreg}")
print(f"The accuracy score for random forest model is {accuracy_random_forest}")
print(f"The accuracy score for grid search random forest model is {accuracy_random_forest_cv}")
print(f"The accuracy score for XGBoost model is {accuracy_xgboost}")
```

## Confusion Matrix

Confusion matix, sometimes known as error matrix, is a specific table layout that allows visualization of performance of a supervised learning model. For a binary classification problem, the confusion matrix is a 2 by 2 matrix whose values are the true positive, false negative, false positive and true negative. 

True positive are those values that are those positive values that are predicted as positive by the model as well. False negative are those values that are predicted to be negative, but are actually positive. Similarly, false positive are those values that are predicted to be positive but are actually negative. Lastly, true negative are those values that those negative values that are predicted as negative by the model. By definition, entry $(i, j)$ in a confusion matrix is the number of observations actually in group $i$, but predicted to be in group
$j$.

We shall now visualize the confusion matrix for the four different models that we have built on the training data (`X_train`).

## ROC Curve and AUC

### ROC Curve (Receiver Operating Characteristic Curve)
The **ROC curve** is a graphical representation of the performance of a binary classification model. It plots the **True Positive Rate (TPR)** against the **False Positive Rate (FPR)** at various threshold values. 

- **True Positive Rate (TPR)**, also known as **Sensitivity** or **Recall**, is the proportion of actual positives correctly identified by the model.
  
  $$
  TPR = \frac{TP}{TP + FN}
  $$

- **False Positive Rate (FPR)** is the proportion of actual negatives incorrectly classified as positives.

  $$
  FPR = \frac{FP}{FP + TN}
  $$
Where $TP$, $FP$, $TN$ and $FN$ indicate the number of true positives, false positives, true negatives and false negatives respectively.

The ROC curve helps to visualize the trade-off between sensitivity (true positive rate) and specificity (1 - false positive rate). As the threshold changes, the TPR and FPR will change, resulting in a curve that shows how well the model performs across all possible thresholds.

### AUC (Area Under the Curve)
The **AUC** stands for **Area Under the ROC Curve**. It is a scalar value between 0 and 1 that quantifies the overall ability of the model to distinguish between positive and negative classes. 

- **AUC = 1** indicates a perfect classifier.
- **AUC = 0.5** indicates a model with no discrimination ability (i.e., random guessing).
- **AUC < 0.5** indicates a model that performs worse than random guessing.

In general, a higher AUC value implies better model performance, as it reflects a model's ability to rank positive instances higher than negative ones across all thresholds.

### Why ROC and AUC Are Useful in Classification
The ROC curve is particularly useful when the classes are imbalanced (i.e., when one class is more frequent than the other). Since the ROC curve is based on the true positive rate and false positive rate, it doesn't rely on the class distribution, making it a robust evaluation tool.

AUC, as a summary statistic, provides a single number to compare multiple models, offering insight into their overall performance. It is especially helpful when comparing models on datasets with imbalanced classes or varying decision thresholds.


```python
from sklearn.metrics import roc_curve, auc, RocCurveDisplay

# Calculate ROC curve for Logistic Regression
fpr1, tpr1, thresholds = roc_curve(y_test, y_pred1)
roc_auc1 = auc(fpr1, tpr1)
display1 = RocCurveDisplay(fpr=fpr1, tpr=tpr1, roc_auc=roc_auc1, name='Logistic Regression')
display1.plot()
plt.show()
```

```python
# Calculate ROC curve for Random Forest Classifier
fpr2, tpr2, thresholds = roc_curve(y_test, y_pred2)
roc_auc2 = auc(fpr2, tpr2)
display2 = RocCurveDisplay(fpr=fpr2, tpr=tpr2, roc_auc=roc_auc2, name='Random Forest Classifier')
display2.plot()
plt.show()
```

```python
# Calculate ROC curve for Random Forest Classifier with Hyperparameter Tuning
fpr3, tpr3, thresholds = roc_curve(y_test, y_pred3)
roc_auc3 = auc(fpr3, tpr3)
display3 = RocCurveDisplay(fpr=fpr3, tpr=tpr3, roc_auc=roc_auc3, name='Random Forest Classifier with Hyperparameter Tuning')
display3.plot()
plt.show()
```

```python
# Calculate ROC curve for XGBClassifier
fpr4, tpr4, thresholds = roc_curve(y_test, y_pred4)
roc_auc4 = auc(fpr4, tpr4)
display4 = RocCurveDisplay(fpr=fpr4, tpr=tpr4, roc_auc=roc_auc4, name='XGBClassifier')
display4.plot()
plt.show()
```
